```python
from PIL import Image
```

```python
im1 = Image.open("NWS_images/cloud_cover/broken/nbkn.png")
h1 = im1.size[0]
w1 = h1/2
```

```python
im2 = Image.open("NWS_images/cloud_cover/few/few.png")
h2 = im2.size[0]
w2 = h2/2
```

```python
box1 = (0, 0, w1, h1)
region1 = im1.crop(box1)
region1
```

```python
box2 = (0, 0, w2, h2)
region2 = im2.crop(box2)
region2
```

```python
if h1 != 86:
    w_new = 43
    h_new = 86
    region1 = region1.resize((w_new, h_new))
else:
    pass
```

```python
def merge(im1, im2):
    w = im1.size[0] + im2.size[0]
    h = max(im1.size[1], im2.size[1])
    im = Image.new("RGBA", (w, h))

    im.paste(im1)
    im.paste(im2, (im1.size[0], 0))

    return im
```

```python
forecast = merge(region1, region2)
forecast
```

```python
img = Image.open('wind_barb.png') 
rgba = img.convert("RGBA") 
datas = rgba.getdata() 
  
newData = [] 
for item in datas: 
    if item[0] == 255 and item[1] == 255 and item[2] == 255:  # finding black colour by its RGB value 
        # storing a transparent value when we find a black colour 
        newData.append((255, 255, 255, 0)) 
    else: 
        newData.append(item)  # other colours remain unchanged 
  
rgba.putdata(newData) 
rgba.save("images/transparent_barb.png", "PNG")
```
