```python
from PIL import Image
```

```python
im1 = Image.open("NWS_images/rain/ra30.png")
im1
```

```python
im2 = Image.open("NWS_images/few/few.png")
im2
```

```python
box1 = (43, 0, 86 ,86)
region1 = im1.crop(box1)
region1
```

```python
box2 = (0, 0, 43, 86)
region2 = im2.crop(box2)
region2
```

```python
def merge(im1, im2):
    w = im1.size[0] + im2.size[0]
    h = max(im1.size[1], im2.size[1])
    im = Image.new("RGBA", (w, h))

    im.paste(im1)
    im.paste(im2, (im1.size[0], 0))

    return im
```

```python
forecast = merge(region1, region2)
forecast
```
